```python
import numpy as np
import PIL
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img= np.array(PIL.Image.open("/content/drive/MyDrive/hills.jpg"))
img= img.sum(axis=-1)
```

```python
def blurring(image):
    for i in range(1,image.shape[0]-1):
        for j in range(1,image.shape[1]-1):
            image[i,j] = (4*image[i,j] + image[i-1,j] + image[i+1,j] + image[i,j-1] + image[i,j+1])/8.0
    return image
```

```python
img_blur=blurring(img)
plt.figure(figsize=(100,100))
plt.imshow(img_blur)
```
