```python
import numpy as np
import PIL
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img= np.array(PIL.Image.open("/content/drive/MyDrive/bees.jpg"))
```

```python
def grayscaling(image):
  return np.dot(image[...,:3],[0.299, 0.587, 0.144])
```

```python
img_gray = grayscaling(img)
image_gray=Image.fromarray(img_gray)
plt.imshow(image_gray)
plt.show()
```
